```python
import pandas as pd
```

#### Data Cleaning

```python
df = pd.read_csv("DataAnalyst.csv")
```

```python
df["Location"].unique()
```

```python
new = df["Location"].str.split(",", n = 1, expand = True) 
```

```python
df["City"] = new[0] 
df["State"] = new[1]
```

```python
#df_CA = df.loc[df["State"] == " CA"]
```

```python
df.head()
```

```python
salary = df["Salary Estimate"].str.split("-", n = 1, expand = True) 
```

```python
df["SalaryEstimate_low"] = salary[0]
df["SalaryEstimate_high"] = salary[1]
a = df["SalaryEstimate_low"].str.split("K", n = 1, expand = True) 
b = df["SalaryEstimate_high"].str.split("K", n = 1, expand = True) 
df["SalaryEstimate_low"] = a[0]
df["SalaryEstimate_high"] = b[0]
c = df["SalaryEstimate_low"].str.split("$", n = 1, expand = True) 
d = df["SalaryEstimate_high"].str.split("$", n = 1, expand = True) 
df["SalaryEstimate_low"] = c[1]
df["SalaryEstimate_high"] = d[1]
df[["SalaryEstimate_low", "SalaryEstimate_high"]] = df[["SalaryEstimate_low", "SalaryEstimate_high"]].apply(pd.to_numeric, errors='coerce')
df[["SalaryEstimate_low","SalaryEstimate_high"]] = df[["SalaryEstimate_low","SalaryEstimate_high"]]*1000
```

```python
df["Revenue"] = df["Revenue"].str.split("billion", n = 1, expand = True) 
revenue = df["Revenue"].str.split("to", n = 1, expand = True) 
df["Revenue_low"] = revenue[0]
df["Revenue_high"] = revenue[1]
x = df["Revenue_high"].str.split(" ", n = 1, expand = True) 
df["Revenue_high"] = x[1]
y = df["Revenue_high"].str.split(" ", n = 1, expand = True)
df["Revenue_high"] = y[0]
z = df["Revenue_low"].str.split("$", n = 1, expand = True) 
t = df["Revenue_high"].str.split("$", n = 1, expand = True) 
df["Revenue_low"] = z[1]
df["Revenue_high"] = t[1]
df[["Revenue_low", "Revenue_high"]] = df[["Revenue_low", "Revenue_high"]].apply(pd.to_numeric, errors='coerce')
df[["Revenue_low","Revenue_high"]] = df[["Revenue_low","Revenue_high"]]*1000000
```

```python
df.head()
```

```python
df = df.drop(columns=["Revenue","Salary Estimate"])
```
