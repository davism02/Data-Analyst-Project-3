```python
danalyst_db = pd.read_csv("Data Analyst.csv")
danalyst_db.head()
```

```python
df = danalyst_db.rename({'Unnamed: 0': 'ID','Job Title': 'Title', 'Salary Estimate': 'Salary', 'Job Description': 'Description', 'Company Name': 'Company', 'Type of ownership': 'ownership','Easy Apply': 'EasyApply'}, axis='columns')
```

```python
df.head()
```

```python
df.tail()
```

```python
df[['Company','Other']] = df.Company.str.split("\n",expand=True)
```

```python
df.head()
```

```python
clean_df = df.drop(['Other', 'Description' , 'EasyApply', 'Competitors'], axis=1)
```

```python
clean_df.head()
```

```python
clean_df['salary']=clean_df['Salary'].str.replace(" (Glassdoor est.)",'',regex=False)
new=clean_df['salary'].str.split("-",n=1,expand=True)
clean_df['min']=new[0]
clean_df['max']=new[1]
```

```python
clean_df.head()
```

```python
clean_df = clean_df.drop(['salary'], axis=1)
```

```python
clean_df.head()
```

```python
#Replace -1 or -1.0 or '-1' to NaN
clean_df=clean_df.replace(-1,np.nan)
clean_df=clean_df.replace(-1.0,np.nan)
clean_df=clean_df.replace('-1',np.nan)
clean_df
```

```python
clean_df.to_csv('clean_data_analyst.csv', index=False)
```
